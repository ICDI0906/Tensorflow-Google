```python
# 使用tensorflow 实现指数学习率
import tensorflow as tf
import matplotlib.pyplot as plt
learning_rate = 0.1
global_step = 1000
decay_step = 100
decay_rate = 0.96

global_ = tf.Variable(tf.constant(0))
x = tf.train.exponential_decay(learning_rate,global_,decay_step,decay_rate,staircase = True)
x1 = tf.train.exponential_decay(learning_rate,global_,decay_step,decay_rate,staircase = False)

T_C = []
T_X = []
with tf.Session() as sess:
    for i in range (global_step):
        T_C.append(sess.run(x,feed_dict = {global_: i}))
        T_X.append(sess.run(x1,feed_dict = {global_: i}))
plt.figure(1)
plt.plot(range(global_step),T_C,'r--')
plt.plot(range(global_step),T_X,'b--')

# decayed_learning_rate = learning_rate *
#                         decay_rate ^ (global_step / decay_steps)
```

### 衰减速率
- $decayed\_learning\_rate = learning\_rate * decay\_rate ^ {\frac{global\_step}{decay\_steps}} $

```python
# 手写指数衰减学习率，当随着运行次数的增加，其学习速率是逐渐减小的
```

```python
import numpy as np
x = np.array(range(1000))
y = learning_rate * np.power(decay_rate, x / decay_step)
y_stair = learning_rate * np.power(decay_rate, list(map(int, x / decay_step)))
```

```python
plt.ylim(0.06,0.1)
plt.plot(x,y,'--',c='b')
plt.plot(x,y_stair,'-','black')
```
